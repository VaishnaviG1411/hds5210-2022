````python
import requests
import bs4

def parse_banner(link):
    headers = { "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36" }
    result = requests.get(link,headers=headers)
    type(result)
    soup = bs4.BeautifulSoup(result.text,"lxml")

    #soup.select('strong') Slecting all elements under 'strong' tag

    #All the states
    states = []

    [states.append(soup.select('strong')[i].getText()) for i in range(0,6)]

    print("State Names : ", states)
    
    soup.select("[rel = 'noopener noreferrer']")
    

    #Arizona list
    arizona = []
    for i in range (0,18):
        arizona.append([soup.select("[rel = 'noopener noreferrer']")[i].getText(),"Arizona"]) 
    #California list
    california = []
    for i in range (18,19):
        california.append([soup.select("[rel = 'noopener noreferrer']")[i].getText(),"California"])
    #Colorado list
    colorado = []
    for i in range (19,24):
        colorado.append([soup.select("[rel = 'noopener noreferrer']")[i].getText(),"Colorado"])
    #Nebraska list
    nebraska = []
    for i in range (24,25):
        nebraska.append([soup.select("[rel = 'noopener noreferrer']")[i].getText(),"Nebraska"]) 
    #Nevada
    nevada = []
    for i in range (25,26):
        nevada.append([soup.select("[rel = 'noopener noreferrer']")[i].getText(),"Nevada"]) 
    #Wyoming
    wyoming = []
    for i in range (26,31):
        wyoming.append([soup.select("[rel = 'noopener noreferrer']")[i].getText(),"Wyoming"]) 
    #['Arizona', 'California', 'Colorado', 'Nebraska', 'Nevada', 'Wyoming']
    result = arizona + california + colorado + nebraska + nevada + wyoming

    return result
````
